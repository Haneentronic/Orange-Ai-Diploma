<h1 style="color:blue">Reshape dataframe using stack/unstack</h1>

```python
import pandas as pd
df = pd.read_excel("stocks.xlsx",header=[0,1])
df
```

```python
x = df 
x = x.stack()
x.to_csv("comapny.csv")
x
```

```python
df.stack(level=0)
```

```python
df_stacked=df.stack()
df_stacked
```

```python
df_stacked.unstack()
```

```python
df_stacked[df_stacked.columns[0]]
```
